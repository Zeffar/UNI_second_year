```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cnt3_array = []
cnt4_array = []
for i in range (1, 5):
    cnt3 = 0
    cnt4 = 0
    for it in range (10000):
        toss = np.random.randint(0, 2, size=10**i)
        has_3hs = np.any(np.convolve(toss, np.ones(3, dtype=int), 'valid') == 3)
        has_4hs = np.any(np.convolve(toss, np.ones(4, dtype=int), 'valid') == 4)
        if has_3hs:
            cnt3 += 1
        if has_4hs:
            cnt4 += 1
    print(f"nb of sequences for {10**i} numbers: cnt3 = {cnt3}, cnt4 = {cnt4}")
    cnt3_array.append(cnt3/10000)
    cnt4_array.append(cnt4/10000)    

plt.plot(range(1, 5), cnt3_array, label='3 consecutive heads')
plt.plot(range(1, 5), cnt4_array, label='4 consecutive heads')
plt.title('Probability of sequencies of k heads in 10k toss sequencies')
plt.xlabel('Power of 10')
plt.ylabel('probability')
plt.legend()
plt.show()
```
