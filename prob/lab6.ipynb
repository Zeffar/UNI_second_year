```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def bernoulli(p):
    return 1 if np.random.random() < p else 0

def simuleaza_binomiala(p, numar_simulari):
    return np.random.binomial(20, p, numar_simulari)
    
def simuleaza_geometrica(p, numar_simulari):
    return np.random.geometric(p, numar_simulari)
```

```python
# Parametri
p = 0.5
numar_simulari = 10000

# Simulare
X_geometrica = simuleaza_geometrica(p, numar_simulari)
X_binomiala = simuleaza_binomiala(p, numar_simulari)


# Histogram
max_val = int(np.max(X_binomiala))
valori = np.arange(1, max_val + 1)
plt.hist(X_binomiala, bins=valori, density=True, alpha=0.75)
plt.xlabel('Valori ale lui X')
plt.ylabel('Frecvena relativ')
plt.title(f'Histogram Binomial(p={p})')
plt.show()


max_val = int(np.max(X_geometrica))
valori = np.arange(1, max_val + 1)
plt.hist(X_geometrica, bins=valori, density=True, alpha=0.75)
plt.xlabel('Valori ale lui X')
plt.ylabel('Frecvena relativ')
plt.title(f'Histogram Geometrica(p={p})')
plt.show()
```

```python
numar_simulari = 10000
valori = [1, 2, 3, 4]
frecvente = [1/2, 1/3, 0, 1/6]  
V = np.random.choice(valori, size=numar_simulari, p=frecvente)

plt.hist(V, bins=np.arange(1, len(valori) + 2) - 0.5, density=True, alpha=0.75)
plt.xlabel('Valori ale lui V')
plt.ylabel('Frecvena relativ')
plt.title('Histogram pentru variabila V')
plt.xticks(valori)
plt.show()

print()

for x in range (1, 5):
    count_x = np.sum(V == x)
    print(f'Numrul de apariii ale valorii {x}: {count_x}')

```
